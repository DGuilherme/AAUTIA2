```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = 'https://raw.githubusercontent.com/DGuilherme/AAUTIA2/main/Doc2Vec/dataset/impi.csv'
dados = pd.read_csv(url,delimiter=r";")
```

```python
#plots
dados['patentes'] = pd.to_numeric(dados['patentes'].str.replace(',', '.'))
dados.plot(x='ano',y='patentes')
```
